
## 1) Generate a numeric dataset

We create 500 rows with 6 columns:  
`[day, item, price, qty, discount, paid]`

- Reproducible seed (`np.random.seed`)  
- Discrete sampling (`np.random.choice`, `np.arange`)  
- Continuous values (`np.random.uniform`)  
- Stack columns (`np.column_stack`)


```python
# Reproducible
np.random.seed(42)
```

```python
# Sizes
N       = 500
n_days  = 30
n_items = 50
```

```python
# Columns
# TODO : Use Random Choice inside of it np.arange with number of days, size is N     #0..29
day      = np.random.choice(np.arange(0, n_days), N)
day
```

```python

# TODO : Use Random Choice inside of it np.arange with number of items, size is N        # 0..49
item     =    np.random.choice(np.arange(0, n_items), N)
item
```

```python

# TODO : Use Random uniform from 5 to 50 with size N                # price (float)
price    = np.random.uniform(5, 50, N)
price
```

```python

# TODO : Use Random Choice inside of it numpy array [1,2,3,4,5], size is N        # quantity (int-like)
qty      =  np.random.choice(np.array([1,2,3,4,5]), N)
qty
```

```python

# TODO : Use Random uniform from 0 to 5 with size N                        # discount (float)
discount = np.random.uniform(0, 5, N)
discount
```

```python

# TODO : Use Random Choice inside of it numpy array [0,1], size is N           # 0=cancel, 1=paid
paid     = np.random.choice(np.array([0,1]), N)
paid
```

```python

# TODO : Assemble design matrix (note: dtype will be float due to float columns) , hint use column stack
X = np.column_stack((day, item, price, qty, discount, paid))
X
```

```python
print("Shape (rows, cols):", X.shape)
print("First 5 rows:")
print(np.round(X[:5], 2))
print("\nColumn order = [day, item, price, qty, discount, paid]")
```


## 3) Simple clean-up using masks

- Detect NaNs with `np.isnan`  
- Impute them with `0.0` using `np.where`
- Update the table `X`


```python

# TODO :  Detect NaNs in price , Hint ## Google isnan method
is_nan_price = np.isnan(price)
print("NaN count in price:", int(is_nan_price.sum()))
```

```python

# TODO : Replace NAN with 0.0 otherwise the price
price = np.where(is_nan_price, 0.0, price)
```

```python
# Update X with the new price
X[:, 2] = price
```

```python

# Quick check
print("After impute  NaN count in price:", int(np.isnan(price).sum()))
```
